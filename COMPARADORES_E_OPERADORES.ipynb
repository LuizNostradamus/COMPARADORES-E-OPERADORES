COMPARADORES E OPERADORES

```python
idade1 = int(input("Digite sua idade: "))
idade2 = int(input("Digite sua idade2: "))
### igual
idade1 == idade2
### diferente
idade1 != idade2
### Maior que
idade1 > idade2
### Menor que
idade1 < idade2
### Maior ou igual
idade1 >= idade2
### Menor ou igual
idade1 <= idade2
print("Idade_1 é igual Idade_2?:", idade1 == idade2)
print("Idade_1 é diferente de Idade_2?:", idade1 != idade2)
print("Idade_1 é maior que Idade_2?:", idade1 > idade2)
print("Idade_1 é menor que Idade_2?:", idade1 < idade2)
print("Idade_1 é maior ou igual a Idade_2?:", idade1 >= idade2)
print("Idade_1 é menor ou igual a Idade_2?:", idade1 <= idade2)
```

```python
### AND PRECISA QUE TODOS SEJAM VERDADEIRAS ###
print("8 é maior que 7 e 7 maior que 8",8 > 7 and 7> 8)
print("8 é maior que 7 e 7 maior que 8: ", 8 > 7 and 7 > 8)
### OR PRECISA APENAS DE UMA VERDADEIRA ###
print("8 é maior que 7 ou 7 menor que 8: ", 8 > 7 or 9 < 8)
print("8 é maior que 7 ou 7 menor que 8: ",7 >= 7 or 10<= 10)
```
